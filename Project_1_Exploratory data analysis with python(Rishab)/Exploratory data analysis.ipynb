```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Here the file was not opened directly without encoding.
#r'string' should use for handle the error for special character
df = pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

Data Cleaning

```python
df.info()
```

```python
#Deleting any column
#Here axis=1 means select row
#inplace = true means save my operation
df.drop(['Status', 'unnamed1'], axis= 1, inplace= True)
```

```python
#Checking Null values
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
#deleting null values
df.dropna(inplace=True)
```

```python
df.shape
```

Changing Datatype

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
#See all column
df.columns
```

```python
#Reanme any column
df.rename(columns={'Marital_Status' : 'Nikah'})
```

Description


```python
df.describe()
```

```python
#Use describe for specific column
df[['Age', 'Orders', 'Amount']].describe()
```

***Exploratory Data Analysis***

Gender

```python
ax = sns.countplot(x = 'Gender',data = df)
```

```python
#For showing values in bar
ax = sns.countplot(x = 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    
sns
    
```

*Data Grouping*

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

*From above graph we can see that most buyers are females and even the purchasing power of females are greater than men*

***Maritial Status***

```python
ax = sns.countplot(x='Marital_Status', hue= 'Gender' ,data= df)
sns.set(rc={'figure.figsize':(2,5)})
df.columns
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(5)
sns.barplot(x='Marital_Status',y='Amount', hue= 'Gender',data = sales_marital)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(6)

sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```
